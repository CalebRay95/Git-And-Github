## ⚙️ First-Time Setup
*Git's first question: "Who are you?" — same energy as meeting your girlfriend's parents.*

```bash
git config --global user.name "Your Name"
git config --global user.email "you@example.com"
git config --global init.defaultBranch main
git config --global core.editor "code --wait"
git config --global push.autoSetupRemote true
git config --list                              # verify everything looks right
```

## 🐣 Start a Repo
*Every great project starts with `git init`. Every disaster also starts with `git init`.*

```bash
git init                          # New local repo — "it begins"
git clone <url>                   # Download existing repo
git clone <url> folder-name       # Clone into specific folder
```

## 🔁 The Daily Loop
*This is your life now. Wake up. Pull. Code. Commit. Push. Repeat until retirement.*

```bash
git status                        # "What did I do?" — ask this 50 times a day
git diff                          # What EXACTLY changed?
git diff --staged                 # What's about to be committed?
git add .                         # Stage everything — the YOLO approach
git add <file>                    # Stage one file — the careful approach
git commit -m "message"           # Save your checkpoint. Not "fix" — an actual message.
git commit -am "message"          # Stage tracked files + commit in one go
git push                          # Upload to GitHub. You're done. You may eat now.
git pull                          # Get teammate's changes before they get mad you didn't
```

## 🕵️ History & Inspection
*Who wrote this??? Use `git blame` and find out. Then realize it was you. Three months ago. At 2am.*

```bash
git log --oneline                    # Short history
git log --oneline --graph --all      # Visual branch tree — poetry in ASCII
git show <hash>                      # One commit details
git blame <file>                     # Who wrote each line (the game of pointing fingers)
git reflog                           # Every HEAD movement ever — your time machine
git shortlog                         # Commits grouped by author
```

## 🌿 Branches
*Branches are like alternate universes. Main is where you live. Feature branches are where experiments go to die or succeed.*

```bash
git branch                        # List local branches  (* = you are here)
git branch -a                     # List ALL branches (local + remote)
git checkout -b <name>            # Create + switch — old but reliable
git switch -c <name>              # Modern create + switch
git switch <name>                 # Modern switch branch
git branch -d <name>              # Delete (safe — only if merged)
git branch -D <name>              # FORCE delete — no questions asked ⚠️
git push origin --delete <name>   # Delete remote branch (the ghost still haunts your PRs)
git branch -vv                    # Show branches + tracking info
```

## 🤝 Merge & Rebase
*Merge: "let's talk about our differences." Rebase: "let's pretend the differences never existed."*

```bash
git merge <branch>                # Merge into current branch
git merge --no-ff <branch>        # Merge with a dedicated merge commit
git merge --abort                 # "I give up" button
git rebase main                   # Rebase onto main — linear history
git rebase -i HEAD~3              # Interactive: clean up last 3 commits before PR
git rebase --abort                # Cancel rebase — walk away slowly
git cherry-pick <hash>            # Pluck one specific commit and apply it here
```

## 💊 Undoing Things
*We don't make mistakes. We make `git reset` opportunities.*

```bash
git restore <file>                # Discard working dir changes — gone forever ⚠️
git restore --staged <file>       # Unstage (chill, file still exists)
git commit --amend -m "msg"       # Fix last commit message — before pushing ONLY
git reset --soft HEAD~1           # Undo commit → changes go back to staging
git reset HEAD~1                  # Undo commit → changes go back to working dir
git reset --hard HEAD~1           # Nuclear option — undo + delete everything ⚠️
git revert <hash>                 # Safe undo — creates new commit (use on shared branches)
```

## 📡 Remote Management
*Remote = the version your teammates can actually see. Keep it clean. Keep it working. They're watching.*

```bash
git remote add origin <url>       # Connect local repo to GitHub
git remote -v                     # List all remotes + their URLs
git remote set-url origin <url>   # Change remote URL (when things move)
git push -u origin main           # First push — sets upstream so future `git push` works
git fetch --all                   # Download all remote branches without merging
git pull --rebase                 # Pull with rebase — cleaner history
git push --force-with-lease       # Safer force push — fails if someone else pushed
```

## 🧳 Stash
*Stash = stuffing your messy work under the bed before your manager does a surprise desk visit.*

```bash
git stash                         # Save changes and clean workspace
git stash push -m "label"         # Save with a name (future you will be grateful)
git stash list                    # See all stashed panic-saves
git stash pop                     # Apply most recent + remove from list
git stash apply stash@{1}         # Apply specific stash (keep in list)
git stash drop                    # Delete top stash
```

## 🧠 Advanced Commands
*The advanced section. Mention these in standup and watch your team nod respectfully.*

```bash
# Bug hunting — the detective work
git bisect start
git bisect bad                    # Current: broken
git bisect good <hash>            # This: was working
git bisect reset                  # Done, go home

# Submodules — repo inside a repo, like inception but for code
git submodule add <url>
git submodule update --init

# Tags — mark the important moments in history
git tag v1.0.0
git tag -a v1.0.0 -m "Release v1"
git push --tags
```

## 🚀 Life-Saving Aliases
*Set these once. Then type 40% less code for the rest of your career. You're welcome.*

```bash
git config --global alias.st 'status'
git config --global alias.co 'checkout'
git config --global alias.cm 'commit -m'
git config --global alias.lg 'log --oneline --graph --decorate --all'
git config --global alias.unstage 'restore --staged'
git config --global alias.last 'log -1 HEAD'

# Now 'git status' → 'git st'. You're basically a hacker now.
```
