## 🪟 Windows — Three Ways, Pick Your Weapon

### Option A — Official Installer ✅ (Recommended for beginners)
1. Go to **https://git-scm.com/download/win** — it auto-downloads
2. Run the `.exe`, click through the wizard
3. **Important settings to choose:**
   - ✅ **Git Bash** — install it (you'll use it constantly)
   - ✅ **"Use Git from the Windows Command line and also from 3rd-party software"**
   - ✅ Line endings: **"Checkout Windows-style, commit Unix-style"** — prevents 10,000 fake changes on your team

### Option B — winget (PowerShell as Admin)
```powershell
winget install --id Git.Git -e --source winget
# One line. Done. No drama. This is the way.
```

### Option C — Chocolatey (if you already use choco)
```powershell
choco install git
```

### Verify it worked (PowerShell, CMD, or Git Bash):
```powershell
git --version
# Expected output: git version 2.4x.x.windows.x
# If nothing shows: restart your terminal. Terminals are dramatic about PATH changes.
```

### Terminals on Windows — Know Your Tools:

| Terminal | Personality | Best for |
|----------|------------|----------|
| **CMD** | The grumpy grandpa — barely functional | Basic stuff, avoid if possible |
| **PowerShell** | Modern, blue, slightly intimidating | Git + scripting + actually useful |
| **Git Bash** | Unix commandline bundled with Git | **Git commands ✅ Use this** |
| **Windows Terminal** | Beautiful tabbed shell | All of the above in one place |

> 💬 Advice from every developer who's suffered through CMD:
> **Use Git Bash on Windows.** Same commands as Mac and Linux.
> No "why does `ls` not work" moments. No unexplained errors.
> Just Unix commands that behave like Unix commands. Revolutionary.

## 🐧 Linux — One Command and You're Done

```bash
# Ubuntu / Debian (most common):
sudo apt update && sudo apt install git -y

# Fedora / RHEL:
sudo dnf install git -y

# Arch / Manjaro (for the people who tell everyone they use Arch):
sudo pacman -S git

# Verify:
git --version
```

> 💬 Linux users installing Git takes 10 seconds.
> Linux users then spend 3 hours customizing their terminal prompt.
> The Git installation was the fastest part of the day.

## 🍎 Mac — You Probably Already Have It (Sort Of)

```bash
# Option A — Xcode CLI Tools (pops a dialog, click Install):
xcode-select --install
# This installs Apple's version of git. Gets the job done.

# Option B — Homebrew (get a newer version):
brew install git
# If you don't have brew: /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"

# Verify:
git --version
```

> 💬 Mac users: your computer already has git because Apple is thoughtful.
> But it's a version from 2 years ago because Apple is also Apple.
> Install via Homebrew for the fresh version.
> Worth it.

## ✅ Platform Terminal Summary

| OS | Best Terminal | Why |
|----|--------------|-----|
| 🪟 Windows | **Git Bash** | Unix commands, no surprises, same as Linux/Mac |
| 🐧 Linux | System terminal (bash/zsh) | Already perfect, nothing to add |
| 🍎 Mac | Terminal or iTerm2 | Native, fast, zsh by default since Catalina |

> All Git commands in this handbook work identically across all platforms unless marked otherwise.
> The exception is some file path differences which will be called out explicitly.
