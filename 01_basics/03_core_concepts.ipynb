## 📖 The Git Dictionary

| Term | Plain English |
|------|---------------|
| **Repository (Repo)** | A folder that Git is tracking — contains all your files + their entire history |
| **Local Repo** | The copy sitting on your laptop. Only you can see it (for now). |
| **Remote Repo** | The copy on GitHub. Accessible from any machine, by anyone you permit. |
| **Working Directory** | Your actual project folder — the files you see and edit in VS Code |
| **Staging Area** | The "waiting room" — files you've queued up but not yet committed |
| **Commit** | A snapshot of your entire project at one moment in time. Like a save in a video game. 🎮 |
| **Branch** | A parallel lane of development. Default: `main`. Feature work: its own branch. |
| **Merge** | Combining one branch into another. Usually fine. Sometimes spectacular. |
| **Remote** | The nickname for a remote repo URL. Default name: `origin` |
| **Push** | Upload your local commits → GitHub (sharing with the world) |
| **Pull** | Download GitHub's changes → your machine (syncing with the world) |
| **Fetch** | Download remote changes WITHOUT merging them. Look before you leap. |
| **Clone** | Download an entire remote repo to your machine for the first time |
| **Fork** | Your own copy of someone ELSE's GitHub repo, on your GitHub account |
| **Pull Request (PR)** | A formal request to merge your branch into another branch. Requires review. |
| **HEAD** | A pointer to where you currently are in history. Usually points to latest commit on your branch. |
| **Origin** | The default nickname for your GitHub remote URL |
| **Upstream** | Usually refers to the original repo you forked from |

> 💬 Vocabulary tip: When your senior says **"just push to origin and open a PR"**
> that translates to: **"upload your branch to GitHub and make a merge request"**
> Now you nod correctly. You're welcome.

## 🔄 The Data Flow — How Code Moves

```
  Working Directory           Staging Area         Local Repo           Remote (GitHub)
  (edit your files)  →add→   (ready?)    →commit→  (saved!)    →push→  (shared!)
         ↑                                                                    |
         |_________________________pull / clone________________________________|
```

Think of it like sending a letter:
- ✏️ **Writing the letter** = editing your code files
- 📬 **Putting it in an envelope** = `git add`
- 🔒 **Sealing and addressing it** = `git commit`
- 📮 **Mailing it** = `git push`

> 💬 You can write 50 drafts (working directory changes) before you even pick an envelope.
> You can put several letters in the envelope (staging multiple files before committing).
> You can write the whole letter in one sitting or over 3 weeks.
> Git doesn't judge your pace. Git just remembers everything.

## 📊 File States in Git

| State | What it means | VS Code indicator |
|-------|--------------|------------------|
| **Untracked** | New file Git has never seen | `U` |
| **Modified** | Changed since last commit | `M` |
| **Staged** | `git add`-ed — ready to commit | `A` or green `M` |
| **Unmodified** | Unchanged since last commit | (nothing — invisible, boring) |
| **Deleted** | File was removed | `D` |
| **Conflicted** | Merge conflict — two versions fighting | `C` or `!` |

> 💬 The enemy state is **Conflicted**.
> It means two people edited the same line and Git is standing in the middle saying
> "I refuse to make this decision, sort this out yourselves."
> Chapter 8 (Merging) covers how to handle this without crying.
