##  Essential Setup — Do This First, Before Anything Else

Works exactly the same on    — run in Git Bash, Terminal, or PowerShell.

```bash
# Your name — this gets stamped on EVERY commit you ever make
git config --global user.name "Your Name"

# Your email — MUST match your GitHub account email (otherwise GitHub can't link commits to your profile)
git config --global user.email "you@example.com"

# Default branch name: 'main' (old Git used 'master' — let's move forward)
git config --global init.defaultBranch main

# Set VS Code as your Git editor (for commit messages, interactive rebase, etc.)
git config --global core.editor "code --wait"
```

>  The "--global" flag means this applies to ALL your repos on this machine.
> Set it once. It lives in `~/.gitconfig`. Done forever.
> Like configuring your phone — you don't redo it every app.

##  Verify & Manage Your Config

```bash
git config --list                       # Show ALL settings
git config user.name                    # Check one specific setting
git config --list --show-origin         # Show settings + which file they're from

# Change a setting: just run it again with the new value
git config --global user.name "New Name"

# Remove a setting (if you have multiple conflicting):
git config --global --unset user.name

# Open your config file in VS Code to edit directly:
git config --global --edit
```

##  Config Scopes — Local vs Global vs System

| Flag | Affects | Stored in |
|------|---------|-----------|
| `--global` | Your user account on this machine | `~/.gitconfig` |
| `--local` | Only this specific repo | `.git/config` inside the repo |
| `--system` | Everyone on this computer | `/etc/gitconfig` (needs admin) |

>  **Use `--global` 99% of the time.** That's it. That's the tip.
> `--local` is for when a repo needs different settings (like a work repo with a work email
> while your personal email is global). `--system` is for sysadmins. Ignore it for now.

##  Useful Extra Configs (Set These and Thank Yourself Later)

```bash
# Make terminal output colorful — absolutely do this
git config --global color.ui auto

# Life-saving shortcuts (aliases) — set once, use thousands of times
git config --global alias.st status
git config --global alias.co checkout
git config --global alias.br branch
git config --global alias.cm 'commit -m'
git config --global alias.lg 'log --oneline --graph --decorate --all'
git config --global alias.unstage 'restore --staged'

# Line ending sanity — prevents thousands of fake "file changed" noise in team repos
git config --global core.autocrlf true    #  Windows — convert CRLF to LF on commit
git config --global core.autocrlf input   #  Mac/Linux — keep LF, don't convert

# Auto-set upstream on first push (Git 2.37+) — stop typing -u every time
git config --global push.autoSetupRemote true
```

>  The aliases are the life-changing part.
> `git status` every 30 seconds becomes `git st` every 30 seconds.
> Still obsessive. But now faster and slightly less embarrassing.
