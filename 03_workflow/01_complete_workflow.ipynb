##  The Big Picture — Where Code Lives and Moves

```
  [GitHub Remote Repo]
        |  ↑
  pull  |  |  push
        ↓  |
  [Local Repo — commits live here]
        |  ↑
commit  |  |  reset
        ↓  |
  [Staging Area — the waiting room]
        |  ↑
    add |  |  restore --staged
        ↓  |
  [Working Directory — your actual files]
```

>  Think of it as a 4-room house:
> **Kitchen** (working directory) = where you cook (write code)
> **Dining table** (staging) = where you plate the food (select what to commit)
> **Photo album** (local repo) = permanent record of every meal you served
> **Food delivery app** (GitHub remote) = where others can order from your kitchen
>
> `git add` = plate the food
> `git commit` = photograph it for the album
> `git push` = list it on the delivery app

##  Scenario A — Starting from an Existing Repo (Most Common)

```bash
# 1. Download the repo to your machine
git clone https://github.com/username/repo-name.git
cd repo-name

# 2. Confirm you're clean
git status
# Should say: nothing to commit, working tree clean (your starting point)

# 3. MAKE YOUR CHANGES 
# (write code, edit files, build things, break things)

# 4. See what you changed
git status           # which files changed
git diff             # what EXACTLY changed, line by line

# 5. Stage what you want in this commit
git add specific-file.txt    # stage one file
git add .                    # stage everything (YOLO mode)

# 6. Commit with a meaningful message
git commit -m "Add user login with JWT authentication"
# NOT: git commit -m "stuff" — your future self will curse at you

# 7. Upload to GitHub
git push origin main
# Done! You're a developer now. 
```

>  `git status` is the most important command.
> Run it before add, after add, before commit.
> Run it when you're confused. Run it when you're confident.
> Run it obsessively. Git status is your therapist.

##  Scenario B — Brand New Project from Scratch

```bash
# 1. Create your project folder and enter it
mkdir my-awesome-project
cd my-awesome-project

# 2. Initialize Git — now Git is watching
git init
# Creates a hidden .git/ folder. Git is now tracking everything in here.

# 3. Create something to commit
echo "# My Awesome Project" > README.md

# 4. First commit
git add .
git commit -m "Initial commit — the beginning of something great (probably)"

# 5. Create a NEW EMPTY repo on GitHub
# Go to github.com → New Repository
# IMPORTANT: Do NOT check "Add README" — you already have one!
# Copy the HTTPS or SSH URL

# 6. Link local to GitHub
git remote add origin https://github.com/username/my-awesome-project.git
git remote -v           # verify the URL looks right

# 7. Rename to main if needed
git branch -M main

# 8. First push — set upstream with -u
git push -u origin main
# -u sets the upstream. After this, just 'git push' works forever.
# Done! Your project is on the internet. 
```

>  Creating a repo on GitHub WITH a README when you already have local commits
> is the most common beginner mistake that causes a "rejected — non-fast-forward" error.
> GitHub wants you to pull first. Your local has commits. GitHub has different commits.
> They don't match. Git panics.
> Solution: create empty repo, add remote, push. Don't let GitHub add files for you.

##  The Day-to-Day Loop

```bash
# Morning ritual (before you write a single line of code):
git pull origin main            # Get everyone else's overnight commits

# ...write your brilliant code...

# End of session (before you close the laptop):
git add .
git commit -m "What you actually did — be specific"
git push                        # Up to GitHub, backed up, shareable

# Sensible daily loop summary:
# PULL FIRST → CODE → ADD → COMMIT → PUSH
```

>  Developers who don't pull first in the morning
> discover their colleagues' changes the hard way — during merge conflicts at 4pm.
> Pull first. Always. Non-negotiable.

##  Navigation Commands (Because Getting Lost is Embarrassing)

```bash
pwd                   # Where am I? (Mac / Linux / Git Bash)
Get-Location          # Where am I? (PowerShell)
ls                    # List files (Mac / Linux / Git Bash)
ls -a                 # List ALL files including hidden (.git folder!)
dir                   # List files (Windows CMD — if you're in that situation)
cd folder-name        # Enter a folder (works everywhere)
cd ..                 # Go up one level (works everywhere)
mkdir folder-name     # Create a folder (works everywhere)
```
