# Chapter 7 — Team Workflow
> *One developer is a genius. Five developers on one repo is where Git earns its paycheck.* 👥

---

> 💬 Imagine a group project where everyone edits the same Word document.
> Now imagine that document is your company's production codebase.
> Now imagine one teammate just pushed directly to main without testing.
>
> This chapter teaches you how to survive that teammate.
> Every team has one. Statistically, you may BE that teammate sometimes.
> We don't judge. We just set up branch protection rules.

## 📜 The Golden Rules of Team Git

*Break these and you'll get a very passive-aggressive Slack message.*

1. **Never push directly to `main`** — every change goes through a branch and a PR. No exceptions.
2. **Always pull before you start work** — your code from yesterday is already out of date somehow.
3. **Keep branches short-lived** — merge often. A branch that lives 3 weeks is a conflict waiting to happen.
4. **Don't rebase shared branches** — only rebase YOUR private branches. Rebasing shared branches rewrites history everyone depends on. People will find you.
5. **Use branch protection rules** — enforce the above automatically, so it's GitHub stopping them, not you.

## 🛡️ Branch Protection Rules (The Automatic Bouncer)

Set this up once on GitHub. It enforces standards so you don't have to be the nagging person.

**Setup:** GitHub repo page → Settings → Branches → Add branch protection rule → Branch name: `main`

**Recommended settings (enable all of these):**
- ✅ **Require a pull request before merging** — no direct pushes to main
- ✅ **Require approvals: 1** — at least one teammate must say LGTM
- ✅ **Require status checks to pass** — tests must be green before merge
- ✅ **Require branches to be up to date before merging** — no merging stale branches
- ✅ **Include administrators** — yes, even you, the repo owner. Especially you.

> 💬 "Include administrators" is the real power move.
> It means even the person who SET UP the rules can't bypass them.
> It's the developer equivalent of eating your own cooking.
> If your rules are good enough for the team, they're good enough for you.

## 🤝 The Team Sync Protocol — Who Pulls When?

```bash
# Every single morning, before you write a single character of code:
git checkout main
git pull origin main          # Sync with what the team did while you slept

# Create your branch for today's work:
git checkout -b feature/your-thing

# ... write feature, commit commits ...

# Before opening a PR (get your branch up to date with main):
git checkout main
git pull origin main
git checkout feature/your-thing
git rebase main               # Put your commits on top of latest main
# Fix any conflicts that arise, then:
git push origin feature/your-thing
# Open PR on GitHub → done
```

> 💬 The "pull before you start" rule sounds obvious.
> You will forget it exactly once, spend 2 hours resolving conflicts,
> and NEVER forget it again.
> That's the learning process. We've all paid the tuition.

## 🔄 The PR Review Cycle Visualized

```
You (Dev A)                    Teammate (Reviewer)             GitHub
  │                                    │                          │
  │── push feature branch ─────────────────────────────────────→ │
  │── open PR with description ─────────────────────────────────→ │
  │                                    │ ← review notification ── │
  │                                    │── leave comments ──────→ │
  │ ← "please fix this" notification ──────────────────────────── │
  │── fix code + push again ───────────────────────────────────→  │
  │                                    │── LGTM ✅ + approve ───→ │
  │── squash & merge ───────────────────────────────────────────→ │
  │── delete branch ────────────────────────────────────────────→ │
  │── git pull origin main (sync up) ──────────────────────────── │
```

> 💬 Code review comments that feel brutal at first are the fastest way to improve.
> The reviewer who writes "this function is doing 3 things, it should do 1"
> is not attacking you. They're teaching you — just aggressively.

## ⚠️ Preventing Conflicts Before They Happen

```bash
# Before opening your PR, sync up:
git fetch origin
git rebase origin/main

# Check how many commits you're behind:
git log HEAD..origin/main --oneline
# No output = you're up to date ✅
# Has output = pull/rebase first, then PR

# Compare your branch vs main before writing the PR description:
git diff main...feature/your-branch
```

## 🍴 Fork vs Branch — When to Use Each

| Situation | Use |
|-----------|-----|
| Your own org's repo, you have push access | **Branch** |
| External open source project, no push access | **Fork** |
| Freelance job, client repo, no direct access | **Fork** |
| Internship, team project with write permissions | **Branch** |

> 💬 Branch = you're invited to the party.
> Fork = you're outside the party, you want to join, and you need to knock on the door via a PR.
