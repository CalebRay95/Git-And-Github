# Commands — Status, Log & Diff
> *Your Git inspection toolkit. These commands change nothing and break nothing.*
> *Run them freely. Run them constantly. Run them for fun.* 

---

>  The three questions a developer asks every day:
> 1. "What's the current state of my code?" → `git status`
> 2. "What happened before I got here?" → `git log`
> 3. "What exactly did I change just now?" → `git diff`
>
> Learn these three commands better than you know your own birthday.
> You'll use them more often.

## `git status` — The Crystal Ball

Run this **constantly**. Before adding. After adding. After committing. When confused. When confident.
It cannot lie to you. It is the most honest command in Git.

```bash
git status
# Shows:
# → modified files (changed but not staged)
# → staged files (git add-ed, ready to commit)
# → untracked files (new files Git has never seen)

git status -s
# Compact mode:
#  M = modified (staged = green M, unstaged = red M)
#  A = newly added (staged)
# ?? = untracked (Git has no idea this file exists yet)
```

>  Running `git status` before `git add .` is like looking at your plate
> before finishing your meal — you check what's left so you know what you're committing.
> Developers who skip `git status` and blindly `git add .` have accidentally committed their
> `.env` file with production database passwords to a public repo.
> The internet does not forget. Rotate your credentials immediately after.

## `git log` — The Time Machine Archive

```bash
git log                              # Full log — very verbose, scroll forever
git log --oneline                    # One line per commit  USE THIS ONE
git log --oneline --graph --decorate --all   # Visual branch tree in ASCII 
git log --author="Name"              # Only one person's commits
git log --since="1 week ago"         # What happened in the last week
git log --grep="keyword"             # Search commit messages for a word
git log -p                           # Every commit + its full diff (heavy)
git log --stat                       # Every commit + file names that changed
git log -5                           # Just the last 5 commits

git show <commit-hash>               # Full details of ONE specific commit
git blame <file>                     # Who wrote each line + when + which commit
git reflog                           # ALL HEAD movements — your accident recovery archive
git shortlog                         # Commits grouped and counted by author
```

>  `git blame` is officially for investigation but is frequently used to find out
> "who wrote this absolute disaster of a function."
> The healthy attitude: blame is for learning context, not for pointing fingers.
> The actual usage: mostly pointing fingers.
> (Often the answer is: past you, 3 months ago, at 1am.)

## `git diff` — The Change Microscope

```bash
git diff
# Changes in your working directory that are NOT yet staged
# (what you've done since the last git add)

git diff --staged
# Changes that ARE staged — what would be committed if you ran git commit RIGHT NOW
# Always run this before committing to check what you're actually committing
# Alias: git diff --cached  (same thing, old name)

git diff <branch1> <branch2>
# Compare two entire branches
# Example: git diff main feature-login

git diff <hash1> <hash2>
# Compare two specific commits
```

>  `git diff --staged` before every commit is professional hygiene.
> Like reading what you wrote before pressing Send on an important email.
> You'd be surprised how many times it shows you staged something you didn't mean to.
