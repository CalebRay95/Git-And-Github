# Commands — Add & Commit
> *Stage your work (add). Lock it in as a snapshot (commit).*
> *One of these is reversible. The other one you should do carefully.* 

---

>  The commit message "fix" appears in approximately 47% of all git repositories worldwide.
> The other 47%: "update", "changes", "wip", "asdf", and "aaaaaa".
> The remaining 6%: actual human beings describing what they did.
>
> Be in the 6%. Your teammates will silently thank you.
> Your future self will thank you out loud, probably emotionally.

## `git add` — Choosing What Goes Into the Commit

```bash
git add <file>        # Stage ONE specific file — surgical precision
git add .             # Stage EVERYTHING in current folder — no survivors
git add -A            # Stage ALL changes including deletions in the whole repo
git add *.js          # Stage all files matching a pattern
git add src/          # Stage an entire folder
git add -p            # Interactive staging — choose WHICH PARTS of each file to stage
                      # (Advanced, but incredibly powerful for clean commits)

# Changed your mind about something you staged?
git restore --staged <file>    # Unstage one file — it goes back to working dir
git restore --staged .         # Unstage EVERYTHING — nuclear unstage
# Old equivalent: git reset HEAD <file>  (still works, just older syntax)
```

>  `git add -p` is the hidden gem nobody teaches beginners.
> It lets you stage specific lines within a file — not the whole file.
> Use case: you fixed a bug AND added a new feature in the same file.
> Stage just the bug fix for one commit, the feature for the next.
> Clean commits. Clean history. Clean conscience.

## `git commit` — Creating the Snapshot

```bash
git commit -m "Your message"
# Rule: start with a VERB — Add, Fix, Update, Remove, Refactor, Implement

git commit -am "Your message"
# Stage ALL modified tracked files AND commit — one command
#  Does NOT include brand new files (untracked)! Use git add first for those.

git commit --amend
# Reopen last commit — change message or add files you forgot
#  ONLY do this BEFORE you push. After pushing, amending rewrites history.
# Rewriting pushed history = your teammates will dislike you.

git commit --amend -m "Better message here"
# Just fix the commit message

git commit --amend --no-edit
# Add forgotten files to last commit — keep the message as-is
```

>  `git commit --amend` is the "send email, immediately realize typo, recall email" button.
> Works perfectly before pushing.
> After pushing... you're rewriting history on a shared branch.
> The `--force` push required after that will cause one of your teammates to lose commits.
> The stand-up conversation the next morning will be awkward.

##  Writing Commit Messages That Aren't Embarrassing

```
#  Please don't:
fix
stuff
ok
changes
update
aaa
FINAL COMMIT
final_FINAL v2 WORKING THIS TIME
lol
debugging

#  Do this instead:
Add user authentication with Google OAuth
Fix: prevent duplicate email addresses on signup
Update README with local setup instructions
Remove deprecated payment gateway endpoints
Refactor: extract cart logic into CartService class

#  Industry standard — Conventional Commits format:
feat(auth): add Google OAuth login flow
fix(cart): resolve price rounding bug on checkout
docs: update API authentication guide
chore: upgrade all dependencies to latest
refactor(user): extract profile logic into service
test: add unit tests for cart calculations
```

>  The Conventional Commits format is used by serious companies.
> `feat:`, `fix:`, `docs:`, `chore:`, `refactor:` — these prefixes let you auto-generate changelogs.
> Also: when something breaks in production and you're looking at git log,
> "Add Google OAuth login" tells you exactly what changed.
> "fix" tells you nothing. It's worse than nothing.
