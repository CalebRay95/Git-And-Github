# Commands — Stash, Tags & Clean
> *The miscellaneous toolkit. Powerful once you know it exists.*
> *Most developers discover these by accident during a crisis.* 🧰

---

> 💬 Scenario: You're halfway through a feature. Your manager walks over.
> "Can you fix that urgent bug on main real quick?"
> You have an unfinished half-working mess all over your working directory.
>
> Option A: Commit the half-working mess and look unprofessional.
> Option B: Delete your work, fix bug, redo your work from memory.
> Option C: `git stash` — hide everything in 5 seconds, fix bug, recover everything.
>
> Option C. Always Option C.

## 🧳 `git stash` — The Panic Button / Emergency Hide

Stash = stuff your in-progress work into a temporary pocket so you can switch context without committing unfinished code.

```bash
git stash                       # Save ALL current changes and clean your working dir
git stash push -m "label"       # Save with a descriptive name (future you will be so grateful)
git stash list                  # See everything you've got stashed
git stash pop                   # Apply most recent stash + remove it from the list
git stash apply                 # Apply most recent stash + KEEP it in the list
git stash apply stash@{2}       # Apply a specific stash by its index number
git stash drop                  # Remove the most recent stash without applying
git stash clear                 # DELETE ALL STASHES ⚠️ — gone forever
git stash branch new-branch     # Create a new branch from a stash (very useful!)
```

> 💬 `git stash` is so fast and so invisible that the manager sees nothing.
> Switch to main, fix bug, push, come back, `git stash pop`.
> Your half-finished feature is back exactly where you left it.
>
> Naming your stashes matters. Stash 3 of "WIP" is useless.
> "WIP: add oauth login — halfway through the token refresh" is priceless.

## 🏷️ `git tag` — Marking the Milestones

Tags = permanent markers for important commits. Used for version releases.

```bash
git tag                              # List all existing tags
git tag v1.0.0                       # Lightweight tag at current commit
git tag -a v1.0.0 -m "Release v1"   # Annotated tag (includes message, author, date) ✅ Preferred
git tag -a v1.0.0 <hash>            # Tag a specific past commit
git push origin v1.0.0              # Push ONE specific tag to GitHub
git push --tags                      # Push ALL local tags to GitHub
git tag -d v1.0.0                   # Delete a tag locally
git push origin --delete v1.0.0     # Delete a tag on GitHub
```

> 💬 Semantic versioning: `v1.2.3`
> First number (1) = Major — breaking changes. Users must update or things break.
> Second number (2) = Minor — new features. Backwards compatible.
> Third number (3) = Patch — bug fixes. Everything still works.
>
> Real world: "just push v2.0.0" = "we changed everything, good luck migrating"
> "just push v1.0.1" = "we fixed that one typo everyone complained about" 

## 🧹 `git clean` & `git rm` — The Delete Commands

```bash
git clean -n          # DRY RUN — preview what WOULD be deleted (ALWAYS do this first!)
git clean -f          # Delete untracked files ⚠️ This is permanent. They're gone.
git clean -fd         # Delete untracked files AND empty folders
git clean -fx         # Delete untracked + .gitignore'd files (nuclear clean)

git rm <file>                # Remove file from disk AND stop tracking it in Git
git rm --cached <file>       # Remove from Git tracking ONLY — keep file on your disk
# Supreme use case:
git rm --cached -r node_modules/   # Stop tracking node_modules (after adding to .gitignore)

git mv old-name.txt new-name.txt   # Rename/move + tell Git about it
```

> 💬 `git clean -n` first. Always. Every time. Without exception.
>
> Someone once ran `git clean -f` instead of `git pull` because they were tired.
> Everything they hadn't committed was gone. Permanently. Including 2 hours of work.
> The `-n` flag is free. It costs you 1 second. Use it.
