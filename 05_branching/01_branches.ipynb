# Chapter 8 — Branching
> *Branches let you experiment without burning down the house.*
> *Main is where you live. Feature branches are where you make a mess.* 🌳

---

> 💬 A developer working directly on `main` is like a surgeon
> doing practice runs on the actual patient.
>
> Branches are the simulation.
> When the simulation succeeds, you merge into the patient.
> When the simulation fails catastrophically, you delete the branch and nobody knows.
>
> This is also how engineering works in general.

## 🤔 Why Do Branches Exist?

- **`main`** = the published, working, potentially live version. This should never be broken.
- **`feature/login`** = your experimental new thing. Break it as much as you need.
- Two developers can work on two separate features without touching each other's code.
- When a feature is done and reviewed → merge back into `main`.

> 💬 Analogy for humans:
> You're writing a novel. `main` is the published copy your readers have.
> A branch is your rough draft. You can cross out paragraphs, rewrite chapters,
> start over three times — and the published copy is untouched.
> When the draft is finally good, you publish (merge).

## 🌿 Branch Commands — The Full Toolkit

```bash
git branch                     # List LOCAL branches  (* = you are currently here)
git branch -a                  # List ALL branches — local AND remote (origin/...)
git branch -r                  # List REMOTE branches only

git branch <name>              # Create a branch (but stay where you are)
git checkout <name>            # Switch to existing branch (classic)
git checkout -b <name>         # Create + immediately switch ← most used, classic

git switch <name>              # Modern: switch to existing branch (Git 2.23+)
git switch -c <name>           # Modern: create + switch ← most used, modern

git branch -M <new-name>       # Rename your CURRENT branch
git branch -m <old> <new>      # Rename any branch by name

git branch -d <name>           # Delete a branch safely — only if it's already merged
git branch -D <name>           # Force delete — even if NOT merged ⚠️ data loss possible
git push origin --delete <name>  # Delete the branch on GitHub (remote)
```

> 💬 `git checkout` vs `git switch`:
> `checkout` is the original command that does 47 different things.
> `switch` was added in Git 2.23 to just... switch branches. One thing.
> Both work. But `switch` is clearer. Use it if your Git is 2.23+.

## 🔄 The Feature Branch Workflow (How Real Work Gets Done)

```bash
# Step 1: Sync with main before creating your branch
git checkout main
git pull origin main            # Grab latest work from the team

# Step 2: Create your feature branch
git checkout -b feature/dark-mode
# (or: git switch -c feature/dark-mode)

# Step 3: Do your work
# edit files, write code, make it work

git add .
git commit -m "feat: implement dark mode toggle"
git add .
git commit -m "feat: add dark mode persistence to localStorage"

# Step 4: Push branch to GitHub
git push origin feature/dark-mode

# Step 5: Open a Pull Request on GitHub (UI) → get reviewed → get merged

# Step 6: Clean up after merge
git checkout main
git pull origin main            # Get the merged version
git branch -d feature/dark-mode  # Delete local branch (already merged, safe)
```

> 💬 Naming conventions that make everyone's life easier:
> `feature/user-authentication` — new feature
> `fix/cart-price-rounding` — bug fix
> `hotfix/security-patch` — urgent fix for production
> `chore/upgrade-dependencies` — maintenance work
>
> When 10 branches show in the PR list, good names tell the story instantly.
