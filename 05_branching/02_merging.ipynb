# Chapter 9 — Merging & Conflicts
> *Bringing branches back together. Usually smooth. Sometimes a war.*
> *The war has rules. This chapter teaches you those rules.* 

---

>  Merging is the moment two developers' work comes together.
> If both worked on different files → smooth, automatic, feels like magic.
> If both edited the same line differently → Git stops, crosses its arms, and says:
> "I refuse to make this decision. YOU decide."
>
> That moment is a CONFLICT.
> It feels terrifying the first time. After the third time it's just Tuesday.

##  Merging Basics

```bash
# Step 1: Switch to the branch you want to merge INTO
git checkout main

# Step 2: Merge the feature branch INTO main
git merge feature-login
# If no conflicts: auto-done, fast-forward or merge commit created

# Merge with an explicit merge commit even if fast-forward is possible:
git merge --no-ff feature-login -m "Merge: add login feature"
# --no-ff = no fast-forward — creates a commit even if not needed
# Useful: preserves the history of the feature branch in the graph

# Squash all the feature branch's commits into one clean commit on main:
git merge --squash feature-login
git commit -m "Add login feature"   # Now hand-write a clean message
# Use this when your feature branch has 15 "wip" commits you don't want in main's history

# Something went wrong? Emergency exit:
git merge --abort
```

>  Rule of thumb:
> `--no-ff` for features (keeps branch history visible)
> `--squash` for when you want one clean "Add X feature" commit in main
> Regular merge (fast-forward) for quick simple merges

##  Resolving Merge Conflicts

A conflict happens when **two branches changed the same line(s) differently** and Git can't auto-decide.

### What a conflict looks like inside your file:
```
<<<<<<< HEAD
Line from YOUR current branch (the one you're merging INTO)
=======
Line from the INCOMING branch (the one you're merging FROM)
>>>>>>> feature-login
```

### How to resolve it step by step:
```bash
# Step 1: Find which files have conflicts
git status
# 'both modified: app.py'  ← these are the battlegrounds

# Step 2: Open the conflicted file — find all <<<, ===, >>> markers
# DECISION TIME: keep your version, keep incoming, or combine both?
# DELETE the markers (<<<, ===, >>>) and keep only the final version you want

# Step 3: Stage the resolved file
git add app.py

# Step 4: Complete the merge
git commit
# Git pre-fills a merge commit message for you

# OR — if you just want out of this mess:
git merge --abort   # Return to pre-merge state as if nothing happened
```

>  VS Code shows conflict markers as highlighted blocks with buttons:
> **Accept Current Change** / **Accept Incoming Change** / **Accept Both Changes**
>
> These are much easier to work with than editing raw markers in a text file.
> Open VS Code. Click the buttons. Feel smart.
>
> The best strategy: talk to the teammate whose change conflicts with yours.
> "Hey, your thing changes line 42. Mine does too. Which approach should we keep?"
> 3-minute conversation. Zero guessing.

##  `git rebase` — Summary

Merge creates a merge commit. Rebase rewrites history to be linear.

```bash
git rebase main               # Move your branch on top of latest main
git rebase -i HEAD~3          # Interactive: edit/squash/drop last 3 commits
git rebase --abort            # Cancel
git rebase --continue         # Continue after resolving a conflict
```

 **Never rebase `main`, `develop`, or any branch your teammates are working on.**

##  `git cherry-pick` — Summary

```bash
git cherry-pick <hash>         # Apply ONE specific commit to current branch
git cherry-pick <h1> <h2>      # Multiple cherry-picks
git cherry-pick --abort        # Bail out
```

When to use: A critical bug fix is stuck on a feature branch and production needs it NOW.
