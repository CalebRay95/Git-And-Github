# Chapter 12 — GitHub: Pull Requests & Forks
> *Where code meets collaboration. The place where reviews happen, bugs get caught,*
> *and occasionally egos get bruised.* 

---

>  A Pull Request is basically saying:
> "I made some changes. I think they're good. Can someone please check my homework?"
>
> A code review is basically saying:
> "I checked your homework. Here are 47 comments."
>
> The goal is NOT to be brutal. The goal is to catch the thing the author
> was too close to to see.
> Be the teammate you wish you had when you submitted your first PR.

##  Pull Requests (PRs) — How Merges Get Reviewed

A PR = formally requesting that your branch gets merged into main (or another branch).

```bash
# Step 1: Push your feature branch to GitHub
git push origin feature-login
```

**Step 2: On GitHub —**
1. A yellow banner appears: "Compare & pull request" → click it
2. Write a **title** (describes what changed)
3. Write a **description** (explains WHY, links to any ticket/issue)
4. Add **reviewers** (teammates who should look at this)
5. Click **Create pull request**

### Merge Options When Approving a PR:

| Option | What it does | Best for |
|--------|-------------|----------|
| **Merge commit** | Creates a merge commit, preserves full history | When you want every commit visible |
| **Squash & merge** | All commits → one clean commit on main | When the branch has WIP/messy commits |
| **Rebase & merge** | Linear history, no merge commit | When you want max cleanliness |

>  Writing `Fixes #42` in your PR description is GitHub magic.
> When the PR merges, GitHub automatically closes Issue #42.
> Your project manager sees the issue close. They're happy. You look efficient.
> Costs you 9 characters. Worth it every time.

##  Forks — Contributing to Projects You Don't Own

A Fork = your own personal copy of someone else's GitHub repo, on YOUR account.
Used when you want to contribute to open source (or a project you don't have push access to).

```bash
# Step 1: Click 'Fork' on GitHub (top right of the repo) — GitHub makes your copy

# Step 2: Clone YOUR fork (not the original):
git clone https://github.com/YOUR-USERNAME/repo-name.git
cd repo-name

# Step 3: Add the original repo as 'upstream' (so you can stay updated):
git remote add upstream https://github.com/ORIGINAL-OWNER/repo-name.git
git remote -v
# origin   = YOUR fork (you push here)
# upstream = ORIGINAL project (you pull updates from here)

# Step 4: Make a branch, make your contribution, push to YOUR fork:
git checkout -b fix/typo-in-readme
git add . && git commit -m "Fix typo in README contributing guide"
git push origin fix/typo-in-readme

# Step 5: Open a PR from your fork to the original repo (GitHub UI)

# Step 6: Keep your fork up to date with original project:
git fetch upstream
git checkout main
git merge upstream/main
git push origin main   # Update your fork on GitHub too
```

>  The open source contribution process feels intimidating.
> "Who am I to submit code to this famous project?"
> You're a developer who found a problem and solved it.
> That's exactly who should submit.
>
> The worst that happens: maintainer says "not this time" and explains why.
> That feedback is free education from someone who knows the codebase.
> Take it. Say thanks. Improve. Try again.
