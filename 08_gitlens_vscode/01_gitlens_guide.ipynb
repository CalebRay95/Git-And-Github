##  Installing GitLens

**GitLens** is a free VS Code extension that supercharges Git inside your editor.
See history, blame, branches, and diffs without ever leaving your code file.

### Install it:
1. Open VS Code
2. Press `Ctrl + Shift + X` (Extensions panel opens)
3. Search: **GitLens**
4. Click **Install** — it's the one by **GitKraken** (it's free)

Or in terminal:
```bash
code --install-extension eamodio.gitlens
```

>  GitLens used to lock most features behind a paywall.
> Then GitLens v14 (2023) made almost everything free.
> If someone tells you "GitLens is mostly paid" — they're working with an old version.
> Update your extension. Get everything for free. Be happy.

##  Feature 1 — Inline Blame (The "Who Did This?!" Button)

Hover over any line of code in VS Code and GitLens shows a floating tooltip:
- **Who** changed this line last
- **When** they changed it (3 days ago, not a date — human-readable)
- **What commit** it was part of — click to see the full diff

You can also turn on **persistent blame** — faded grey text at the end of each line.

```
Ctrl + Shift + G  then  B   → Toggle file blame view
```

Terminal equivalent (nowhere near as convenient):
```bash
git blame filename.py    # Output in terminal, line by line, raw
```

>  Using `git blame` in the terminal produces an output that looks like
> a spreadsheet someone designed in 1997.
> GitLens shows the same information as a clean, hoverable overlay.
>
> The feature is called "blame."
> The culture around it should NOT be blame.
> It's context. "I see Sarah changed this 3 days ago — let me ask her why" is the right use.
> Not "AHA SARAH THIS WAS YOUR BUG." (even if it was Sarah's bug)

##  Feature 2 — Source Control Panel (Stage Without Touching Terminal)

Click the **branch icon** in the left VS Code sidebar, or press `Ctrl + Shift + G`.

What you see:
- **Changes** — all modified files with M (modified) / U (untracked) / D (deleted) icons
- **Staged Changes** — files that have been `git add`-ed
- **Commits** — recent commit history
- **Branches** — all local branches

### Stage and commit without opening a terminal:
1. See a file under **Changes**
2. Click the **`+`** next to a file → that file is now staged (`git add` equivalent)
3. Or click **`+`** next to "Changes" header → stages everything
4. Type your commit message in the box at the top
5. Press `Ctrl + Enter` → commit!

>  This is how most developers actually use Git day-to-day.
> Terminal for complex stuff. VS Code Source Control panel for daily staging and committing.
> Your terminal knowledge still matters — but you don't type `git add` 30 times a day.

##  Feature 3 — Commit Graph (Free Since GitLens v14!)

The Commit Graph is a **visual, interactive timeline** of your entire branch history.

>  **GitLens v14 (released 2023) made the Commit Graph completely free.**
> If anyone says "that's a paid feature" — update your GitLens. You're behind.

### How to open it:
- `Ctrl + Shift + P` → type **"GitLens: Show Commit Graph"** → Enter
- OR: GitLens side panel → Commits section → graph icon at the top

### What you actually see:
- Multi-lane visual timeline — each branch gets its own lane
- Every commit as a dot — click to see: files changed, full diff, copy hash
- Branch labels float at the latest commit for each branch
- Merge points, diverge points, where branches split — all visible instantly
- Search by message, author, file, or hash

Terminal equivalent (you'll understand why the GUI wins):
```bash
git log --oneline --graph --decorate --all
# ABCD (main)
#     \
#      EFG (feature/login)
# It works. But the GUI is 1000% easier to read.
```

>  The Commit Graph is the answer to the question
> "what is the actual state of my repo right now?"
> Every branch, every commit, every relationship — visible in 2 seconds.
> Seniors visualize repos like this in their heads.
> You get a GUI that does it for you. Use it.

##  Feature 4 — Branch Tracking (Ahead/Behind)

**Branch tracking** = your local branch knows about its remote counterpart and can tell you the gap.

```bash
# Link local branch to remote when pushing first time:
git push -u origin main
# -u = --set-upstream. Now 'main' tracks 'origin/main'

# Check tracking status:
git status
# "Your branch is ahead of 'origin/main' by 2 commits" → you need to push
# "Your branch is behind 'origin/main' by 3 commits" → you need to pull
# Both at once → diverged, needs rebase or merge

# See all branches with tracking info:
git branch -vv
# * main   a3f8c2d [origin/main: ahead 2] Add login
# [origin/main: ahead 2]  → 2 commits to push
# [origin/main: behind 3] → 3 commits to pull
# [origin/main: gone]     → remote branch was deleted
```

### In GitLens UI:
- Bottom-left status bar in VS Code → you see: `main ↑2 ↓3`
- `↑2` = 2 commits to push
- `↓3` = 3 commits to pull
- Click it to switch branches, publish branches, or sync

>  The `↑↓` indicator in the bottom bar is the first thing you check every morning.
> If `↓` is more than 0: pull before you write a single line of code.
> If `↑` is climbing past 10: push before your laptop catches fire.

##  Feature 5 — File History & Timeline

Right-click any file in VS Code → **GitLens: Open File History**

Or the **Timeline** panel at the bottom of the Explorer sidebar:
- Shows every commit that ever touched that specific file
- Click any entry → see the exact diff for that commit
- Jump to "when did this function start existing?" instantly

Terminal equivalent:
```bash
git log --oneline -- filename.py    # Commits that touched this file
git show <hash>                     # Full diff of one commit
```

##  GitLens Quick Reference

| What you want | How to get it | What it shows |
|---------------|--------------|---------------|
| Who changed this line? | Hover over any line | Author, date, commit |
| Stage a file | Source Control `+` button | Equivalent to `git add` |
| Make a commit | Source Control message box + Ctrl+Enter | `git commit` |
| All branches + status | GitLens side panel star icon | Branches with tracking |
| Visual history | GitLens > Commit Graph | Every commit, every branch |
| File history | Right-click → File History | All commits for that file |
| Push/pull status | Bottom status bar | `↑` push count, `↓` pull count |

>  GitLens doesn't replace learning terminal Git.
> It helps you SEE what Git is doing while your mental model is still forming.
> Learn the commands. Use GitLens to verify your understanding.
> Both together = you actually understand what's happening.
