##  Aliases That Will Save Hours of Your Life

```bash
# Set these once. Use them every day forever.
git config --global alias.st 'status'
git config --global alias.co 'checkout'
git config --global alias.br 'branch'
git config --global alias.cm 'commit -m'
git config --global alias.lg 'log --oneline --graph --decorate --all'
git config --global alias.unstage 'restore --staged'
git config --global alias.last 'log -1 HEAD'

# Usage:
git st          # Full name: git status
git lg          # Beautiful visual branch tree — use this CONSTANTLY
git cm "Fix"    # git commit -m "Fix"
git unstage .   # git restore --staged .
git last        # See exactly what your last commit was
```

>  The `git lg` alias is the sleeper hit.
> `git log --oneline --graph --decorate --all` in full is 40 keystrokes.
> `git lg` is 5 keystrokes.
> Same output. Visual branch tree. Beautiful.
> Set this alias before you set up anything else.

##  Branching Strategies — GitHub Flow vs GitFlow

### GitHub Flow — Best for Most Projects (Simple and Effective)
```
main                ← Always deployable. Protected. Sacred ground.
feature/my-thing    ← Branch from main, PR back to main when ready.
fix/urgent-bug      ← Same pattern. PR → main ASAP.
```
Rule: If it's done and reviewed → it's on main. Main deploys to production.

### Git Flow — Larger Enterprise / Release Cycles
```
main       ← Production only. Tagged releases only go here.
develop    ← Integration branch. Features merge here first.
feature/*  ← New features. Branch from develop.
release/*  ← Staging. Final checks before main.
hotfix/*   ← Emergency production fix. Branch from main. Merge to both main AND develop.
```
More structure. More process. More for teams doing scheduled releases.

>  GitHub Flow is what most startups and modern dev teams use.
> GitFlow is what companies with 4-week release cycles use.
> GitHub Flow = fast cars. GitFlow = carefully assembled spacecraft.
> Pick the right tool for your team's cadence.

##  Your Perfect `.gitconfig` File

View and edit it directly:
```bash
git config --global --edit
```

```ini
[user]
    name = Your Name
    email = you@example.com

[core]
    editor = code --wait
    autocrlf = true    # Windows only. Mac/Linux: use 'input'

[init]
    defaultBranch = main

[push]
    autoSetupRemote = true    # No more typing -u on first push

[alias]
    st = status
    co = checkout
    br = branch
    cm = commit -m
    lg = log --oneline --graph --decorate --all
    unstage = restore --staged
    last = log -1 HEAD

[color]
    ui = auto
```

>  This config file is the single most impactful thing you can set up.
> One config. Every repo on this machine benefits.
> Future you will find this file and be grateful.
> Past you when setting up a new machine will curse that you didn't back it up.

##  The 10 Golden Rules of Git

*Break these and suffer. Follow these and prosper.*

1. **Commit early, commit often.** Small commits are easier to understand, review, and revert. "I have one massive commit from today" is not a safety net.

2. **Never force-push `main`, `develop`, or any shared branch.** Ever. Your teammates' local repos depend on that history. You rewrite it, they suffer.

3. **Pull before you push.** Every single time. Without exception. Sync first, code second.

4. **Write meaningful commit messages.** Your commit message is a note to your future self. "fix" is not a note. Your 3am self debugging prod will thank you or curse you.

5. **Use branches for everything.** Never code directly on `main`. Main is the published book. Branches are your drafts.

6. **Review what you're staging.** `git diff --staged` before every commit. Do you KNOW what you're about to commit? Looking at the diff takes 10 seconds. Reverting a bad commit takes 10 minutes.

7. **Never commit `.env` files, API keys, or passwords.** Put them in `.gitignore` on Day 1. Rotating leaked credentials after they've been public for 3 hours is not fun. I know someone who's done this. It was me.

8. **`git status` is your best friend.** Run it constantly. Obsessively. It cannot lie and cannot betray you.

9. **When in doubt, `git reflog`.** You can almost always recover. The real question is whether you committed first. See Rule 1.

10. **Read the error messages.** Git's error messages are actually really good. They often tell you exactly what went wrong AND what to do. Reading them carefully is underrated.

>  Rule 10 is the most underrated.
> "git says something, I just force-push" is how production incidents happen.
> Read the error. Follow the hint. The computer is trying to help you.

##  Recommended Learning Timeline

```
Week 1 focus: init, clone, add, commit, push, pull, status, log
  → Goal: complete the daily loop without googling

Week 2 focus: branching, merging, conflict resolution, PR workflow
  → Goal: survive working in a team repo

Week 3 focus: reset, revert, stash, reflog
  → Goal: recover from mistakes confidently

Week 4 focus: rebase -i, cherry-pick, tags, aliases, hooks
  → Goal: write clean history, automate standards

Month 2+: Contribute to an open source project
  → Goal: apply everything on a real codebase you didn't write
```
